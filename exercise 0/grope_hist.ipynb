### Generating the  histogramms for the grope test samples

```python
import numpy as np
import matplotlib.pyplot as plt

# read in the data
h=np.loadtxt("Hadron.txt")
e=np.loadtxt("Electron.txt")
m=np.loadtxt("Muon.txt")
t=np.loadtxt("Tau.txt")
h=h.transpose();
e=e.transpose();
m=m.transpose();
t=t.transpose();

# make the histograms
plt.hist(h[2],bins=30,label=r'$Z^0$ $\to$ hadrons')
plt.hist(e[2],bins=30,label=r'$Z^0$ $\to$ $e^+e^-$')
plt.hist(m[2],bins=30,label=r'$Z^0$ $\to$ $\mu^+\mu^-$')
plt.hist(t[2],bins=30,label=r'$Z^0$ $\to$ $\tau^+\tau^-$')
plt.title(r'PCHARGED: histogram of scalar sum of charged momenta')
plt.legend()
plt.xlabel(r'total scalar momentum [GeV]')
plt.show()

plt.hist(h[3],bins=30,label=r'$Z^0$ $\to$ hadrons')
plt.hist(e[3],bins=30,label=r'$Z^0$ $\to$ $e^+e^-$')
plt.hist(m[3],bins=30,label=r'$Z^0$ $\to$ $\mu^+\mu^-$')
plt.hist(t[3],bins=30,label=r'$Z^0$ $\to$ $\tau^+\tau^-$')
plt.title(r'NCHARGED: histogram of charged particles multiplicity')
plt.legend()
plt.xlabel(r'Multiplicity')
plt.show()

plt.hist(h[4],bins=30,label=r'$Z^0$ $\to$ hadrons')
plt.hist(e[4],bins=30,label=r'$Z^0$ $\to$ $e^+e^-$')
plt.hist(m[4],bins=30,label=r'$Z^0$ $\to$ $\mu^+\mu^-$')
plt.hist(t[4],bins=30,label=r'$Z^0$ $\to$ $\tau^+\tau^-$')
plt.title(r'E_CAL: histogram of total energy in the Ecal')
plt.legend()
plt.xlabel(r'total energy [GeV]')
plt.show()

plt.hist(h[5],bins=50,label=r'$Z^0$ $\to$ hadrons')
plt.hist(e[5],bins=30,label=r'$Z^0$ $\to$ $e^+e^-$')
plt.hist(m[5],bins=30,label=r'$Z^0$ $\to$ $\mu^+\mu^-$')
plt.hist(t[5],bins=30,label=r'$Z^0$ $\to$ $\tau^+\tau^-$')
plt.title(r'H_CAL: histogram of total energy in the Hcal')
plt.legend()
plt.xlabel(r'total energy [GeV]')
plt.show()
```
