## Selection of the cuts and efficiency matrix

From the obtained histograms we can try different sets of cuts to get an first identification of the decay channels.
For the ECAL histogram, and after some tries, select a set of 3 values of energy cuts.

The efficiency $\varepsilon$ is defined as the probability of getting a "positive signal" upon the ocurrence of a process.
In the case of the MC samples, we have an amount of 'n' events that pass some selection (defined by the cuts) over a total of N events. This process follow a binomial distribution.

$$ \varepsilon = \frac{n}{N} $$
The error (or uncertainty) can be caluclated as the standad deviation of this distribution, which is given by: the $$ \sigma = \sqrt{\frac{\varepsilon(1-\varepsilon)}{N}} = \sqrt{\frac{n(N-n)}{N^3}} $$


```python

cut1 = 9.0
cut2 = 46.5
cut3 = 78.0

mymask_ee = bran_ee['E_ecal']
mymask_mm = bran_mm['E_ecal'] 
mymask_tt = bran_tt['E_ecal']
mymask_qq = bran_qq['E_ecal']

myeffi_ee_ee = sum(bran_ee['E_ecal'] > cut3)/len(mymask_ee)            # 11
myeffi_mm_ee = sum(bran_mm['E_ecal'] > cut3)/len(mymask_mm)            # 12
myeffi_tt_ee = sum(bran_tt['E_ecal'] > cut3)/len(mymask_tt)            # 13
myeffi_qq_ee = sum(bran_qq['E_ecal'] > cut3)/len(mymask_qq)            # 14

err_ee_ee = np.sqrt(myeffi_ee_ee*(1-myeffi_ee_ee)/len(mymask_ee))
err_mm_ee = np.sqrt(myeffi_mm_ee*(1-myeffi_mm_ee)/len(mymask_mm))
err_tt_ee = np.sqrt(myeffi_tt_ee*(1-myeffi_tt_ee)/len(mymask_tt))
err_qq_ee = np.sqrt(myeffi_qq_ee*(1-myeffi_qq_ee)/len(mymask_qq))

myeffi_ee_mm = sum(bran_ee['E_ecal'] < cut1)/len(mymask_ee)            # 21
myeffi_mm_mm = sum(bran_mm['E_ecal'] < cut1)/len(mymask_mm)            # 22
myeffi_tt_mm = sum(bran_tt['E_ecal'] < cut1)/len(mymask_tt)            # 23
myeffi_qq_mm = sum(bran_qq['E_ecal'] < cut1)/len(mymask_qq)            # 24

err_ee_mm = np.sqrt(myeffi_ee_mm*(1-myeffi_ee_mm)/len(mymask_ee))
err_mm_mm = np.sqrt(myeffi_mm_mm*(1-myeffi_mm_mm)/len(mymask_mm))
err_tt_mm = np.sqrt(myeffi_tt_mm*(1-myeffi_tt_mm)/len(mymask_tt))
err_qq_mm = np.sqrt(myeffi_qq_mm*(1-myeffi_qq_mm)/len(mymask_qq))

myeffi_ee_tt = ( sum(bran_ee['E_ecal']<cut2) - sum(bran_ee['E_ecal']<cut1) )/len(mymask_ee)    # 31
myeffi_mm_tt = ( sum(bran_mm['E_ecal']<cut2) - sum(bran_mm['E_ecal']<cut1) )/len(mymask_mm)    # 32
myeffi_tt_tt = ( sum(bran_tt['E_ecal']<cut2) - sum(bran_tt['E_ecal']<cut1) )/len(mymask_tt)    # 33
myeffi_qq_tt = ( sum(bran_qq['E_ecal']<cut2) - sum(bran_qq['E_ecal']<cut1) )/len(mymask_qq)    # 34

err_ee_tt = np.sqrt(myeffi_ee_tt*(1-myeffi_ee_tt)/len(mymask_ee))
err_mm_tt = np.sqrt(myeffi_mm_tt*(1-myeffi_mm_tt)/len(mymask_mm))
err_tt_tt = np.sqrt(myeffi_tt_tt*(1-myeffi_tt_tt)/len(mymask_tt))
err_qq_tt = np.sqrt(myeffi_qq_tt*(1-myeffi_qq_tt)/len(mymask_qq))

myeffi_ee_qq = ( sum(bran_ee['E_ecal']<cut3) - sum(bran_ee['E_ecal']<cut2) )/len(mymask_ee)    # 41
myeffi_mm_qq = ( sum(bran_mm['E_ecal']<cut3) - sum(bran_mm['E_ecal']<cut2) )/len(mymask_mm)    # 42
myeffi_tt_qq = ( sum(bran_tt['E_ecal']<cut3) - sum(bran_tt['E_ecal']<cut2) )/len(mymask_tt)    # 43
myeffi_qq_qq = ( sum(bran_qq['E_ecal']<cut3) - sum(bran_qq['E_ecal']<cut2) )/len(mymask_qq)    # 44

err_ee_qq = np.sqrt(myeffi_ee_qq*(1-myeffi_ee_qq)/len(mymask_ee))
err_mm_qq = np.sqrt(myeffi_mm_qq*(1-myeffi_mm_qq)/len(mymask_mm))
err_tt_qq = np.sqrt(myeffi_tt_qq*(1-myeffi_tt_qq)/len(mymask_tt))
err_qq_qq = np.sqrt(myeffi_qq_qq*(1-myeffi_qq_qq)/len(mymask_qq))

#print(round(myeffi_ee_ee, 3), round(myeffi_mm_ee, 3), round(myeffi_tt_ee, 3), round(myeffi_qq_ee, 3))
#print(round(myeffi_ee_mm, 3), round(myeffi_mm_mm, 3), round(myeffi_tt_mm, 3), round(myeffi_qq_mm, 3))
#print(round(myeffi_ee_tt, 3), round(myeffi_mm_tt, 3), round(myeffi_tt_tt, 3), round(myeffi_qq_tt, 3))
#print(round(myeffi_ee_qq, 3), round(myeffi_mm_qq, 3), round(myeffi_tt_qq, 3), round(myeffi_qq_qq, 3))
#
#print(round(err_ee_ee, 3), round(err_mm_ee, 3), round(err_tt_ee, 3), round(err_qq_ee, 3))
#print(round(err_ee_mm, 3), round(err_mm_mm, 3), round(err_tt_mm, 3), round(err_qq_mm, 3))
#print(round(err_ee_tt, 3), round(err_mm_tt, 3), round(err_tt_tt, 3), round(err_qq_tt, 3))
#print(round(err_ee_qq, 3), round(err_mm_qq, 3), round(err_tt_qq, 3), round(err_qq_qq, 3))

M_efi = np.array([[myeffi_ee_ee, myeffi_mm_ee, myeffi_tt_ee, myeffi_qq_ee], [myeffi_ee_mm, myeffi_mm_mm, myeffi_tt_mm, myeffi_qq_mm], [myeffi_ee_tt, myeffi_mm_tt, myeffi_tt_tt, myeffi_qq_tt], [myeffi_ee_qq, myeffi_mm_qq, myeffi_tt_qq, myeffi_qq_qq]])

M_err = np.array([[err_ee_ee, err_mm_ee, err_tt_ee, err_qq_ee], [err_ee_mm, err_mm_mm, err_tt_mm, err_qq_mm], [err_ee_tt, err_mm_tt, err_tt_tt, err_qq_tt], [err_ee_qq, err_mm_qq, err_tt_qq, err_qq_qq]])

np.set_printoptions(precision=3, suppress=True)
print(M_efi)
print(M_err)
```

# Exercise 2 - Separate t- and s-channel contributions

As shown in the graph below, the histogram for the MC data of the variable 'cos_thet', where $\theta$ is the polar angle between the incoming positron and outgoing positive particle, in the case of the decay channel $e^+ e^- \rightarrow e^+ e^-$, exhibits a particular behaviour with respect to the others.

```python
from scipy.optimize import curve_fit
import scipy.integrate as integrate

cthet_ee = ak.to_numpy(bran_ee[var[6]])
cthet_mm = ak.to_numpy(bran_mm[var[6]])
cthet_tt = ak.to_numpy(bran_tt[var[6]])
cthet_qq = ak.to_numpy(bran_qq[var[6]])

fig, (ax) = plt.subplots(nrows=1, ncols=1)

ax.hist(cthet_ee, bins=1000, range=(-1.25, 1.25), histtype='step',  linewidth=1, edgecolor='b', label='ee')
ax.hist(cthet_mm, bins=1000, range=(-1.25, 1.25), histtype='step',  linewidth=1, edgecolor='y', label='mm')
ax.hist(cthet_tt, bins=1000, range=(-1.25, 1.25), histtype='step',  linewidth=1, edgecolor='g', label='tt')
ax.hist(cthet_qq, bins=1000, range=(-1.25, 1.25), histtype='step',  linewidth=1, edgecolor='r', label='qq')
ax.set_ylabel('Number of events')
ax.set_xlabel('cos_theta Histogram')
ax.legend()

plt.show()
```

This is due to this decay contains a mixture of s-and t-channels.
We know that the production cross section for the s-channel is given by:

s-channel: $ \frac{d\sigma_s}{d\Omega} \propto 1+\cos^2\theta $

And for the t-channel is:

t-channel: $ \frac{d\sigma_t}{d\Omega} \propto (1-\cos\theta)^{-2} $

Now we select only the decay channel $ee$, where the t-channel is present, and create the function "cross_fun", which is a general function with the combination of the production cross section of both channels. Here A, B and C, are constans to be found. Then we can fit the data of the histogram to separate the contributions.

```python
plt.style.use(mplhep.style.ATLAS)
plt.figure(figsize=(7,5))

x = np.linspace(-0.9, 0.9, 1001)
new_bin_content, new_bin_edges, _ = plt.hist(cthet_ee, x, histtype='step', linewidth=1, label='cos_thet')
x_mid = 0.5*(new_bin_edges[1:] + new_bin_edges[:-1]) #Calculate midpoint of the bars

def cross_fun(x, A, B, C):
    return A*(1 + x**2) + B*((1-x)**(-2)) + C
 
coeff, var_matrix = curve_fit(cross_fun, x_mid, new_bin_content) # coeff: Return fit parameters A & B (array 2x1 in this case), and var_matrix: Estimated covariance of coeff (array 2x2 - diagonals!)
(A,B,C) = coeff

# Get the fitted curve
plt.plot(x_mid, coeff[0]*(1 + x_mid**2), label="s-channel")
plt.plot(x_mid, coeff[1]*(1 - x_mid)**(-2), label="t-channel")
hist_fit = cross_fun(x_mid, *coeff)
plt.plot(x_mid, hist_fit, label = 'Both channels')

plt.title('s- and t-channels')
plt.xlabel('Cos(theta)')
plt.ylabel('Number of events')
plt.legend()
plt.grid()
plt.show()
```

```python
def ch_s(x):
    return A*(1+x**2) + C
def ch_t(x):
    return B*(1-x)**(-2) + C

#Total area of s-channel 
TotA_s = integrate.quad(ch_s,-1,1)[0]
# Partial area of s-channel dominant
ParA_s = integrate.quad(ch_s,-1,0.8)[0]
#Fraction of the area s-channel
Frac = ParA_s/TotA_s
print(Frac)

#Area of t-channel dominant
Area_t = integrate.quad(ch_s, 0.8,1.)[0]
```
