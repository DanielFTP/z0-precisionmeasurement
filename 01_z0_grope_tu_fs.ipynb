# Exercise 0: First Analysis of Grope data
In this first analysis the observed and manually noted data from the Grope simulator will be analysed statistically

```python
'''Importing required libraries'''
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''reading csv input data to pandas df'''
path_to_base_dir = 'Data/'
path_to_output_dir = path_to_base_dir+'output/'

hadron_df = pd.read_csv(path_to_base_dir+'Z_experiment-hadron.csv')
electron_df = pd.read_csv(path_to_base_dir+'Z_experiment-elektron.csv')
myon_df = pd.read_csv(path_to_base_dir+'Z_experiment-myon.csv')
tauon_df = pd.read_csv(path_to_base_dir+'Z_experiment-tauon.csv')


'''Define strings to select parts of the df later
All column names may be viewed with the below command:'''
# print(hadron_df.columns)

ecal_sum_e_string = 'ECAL(SumE)'
hcal_sum_e_string = 'HCAL(SumE)'
Pcharged_string = 'Ctrk(SumP)'
Ncharged_string = 'Ctrk(N)'
```

```python
'''plot histograms for 4 variables for all four particles, 
use the strings from above to select columns to be displayed'''

plt.hist(hadron_df[ecal_sum_e_string], histtype='step', bins=20, color='C1', label='hadron')
plt.hist(electron_df[ecal_sum_e_string], histtype='step',bins=20, color='C2', label='electron')
plt.hist(myon_df[ecal_sum_e_string], histtype='step',bins=20, color='C3', label='muon')
plt.hist(tauon_df[ecal_sum_e_string], histtype='step',bins=20, color='C4', label='tauon')
plt.title('GROPE Data - E_ecal')
plt.ylabel('Counts')
plt.xlabel('E_ecal [GeV]')
plt.legend()
plt.savefig(path_to_output_dir+ 'grope_ecal.pdf')
plt.show()

plt.hist(hadron_df[hcal_sum_e_string], histtype='step', bins=10, range=(0,50), color='C1', label='hadron')
plt.hist(electron_df[hcal_sum_e_string], histtype='step',bins=4, color='C2', label='electron')
plt.hist(myon_df[hcal_sum_e_string], histtype='step',bins=20, color='C3', label='muon')
plt.hist(tauon_df[hcal_sum_e_string], histtype='step',bins=20, color='C4', label='tauon')
plt.title('GROPE Data - E_hcal')
plt.ylabel('Counts')
plt.xlabel('E_hcal [GeV]')
plt.legend()
plt.savefig(path_to_output_dir+ 'grope_hcal.pdf')
plt.show()

plt.hist(hadron_df[Pcharged_string], histtype='step', bins=10,range=(0,110), color='C1', label='hadron')
plt.hist(electron_df[Pcharged_string], histtype='step',bins=4, color='C2', label='electron')
plt.hist(myon_df[Pcharged_string], histtype='step',bins=20, color='C3', label='muon')
plt.hist(tauon_df[Pcharged_string], histtype='step',bins=20, color='C4', label='tauon')
plt.title('GROPE Data - Pcharged')
plt.ylabel('Counts')
plt.xlabel('Pcharged [GeV]')
plt.legend()
plt.savefig(path_to_output_dir+ 'grope_pcharged.pdf')
plt.show()

plt.hist(hadron_df[Ncharged_string], histtype='step', bins=10, range=(0,40), color='C1', label='hadron')
plt.hist(electron_df[Ncharged_string], histtype='step',bins=4, color='C2', label='electron')
plt.hist(myon_df[Ncharged_string], histtype='step',bins=20, color='C3', label='muon')
plt.hist(tauon_df[Ncharged_string], histtype='step',bins=20, color='C4', label='tauon')
plt.title('GROPE Data - Ncharged')
plt.ylabel('Counts')
plt.xlabel('Ncharged')
plt.legend()
plt.savefig(path_to_output_dir+ 'grope_ncharged.pdf')
plt.show()
```

Comment on the binsize:
The number of bins was chosen in a way that the distributions were best visible.
However, due to the very small sample range this information is not very reliable data.

## Initial comments/ guesses for the cuts:
##### Muon
Muons may be completely seperated by cutting the E_cal channel: $0 \leq $ E_cal $\leq 15$.
Further cuts were not necessary.
##### Hadrons
Hadrons may be seperated by cutting the Ncharged channel: $15\leq$ Ncharged $\leq$ open. With the exception of one hadronic event, this perfectly cuts out all other events.


For electrons and tauons no perfect seperation is possible. Thus we suggest the following cuts for each variable to have the best possible seperation between Electrons and Tauons.
Below a table with all 4 parameters can be found:

|       | Electrons   | Tauons      |
|---------|-------------|-------------|
|E_ecal   | 75 to no cut| 15 to 91    |
|E_hcal   | 0 to 7      | 0 to open   |
|Pcharged| 0 to 85| 0 to 100    |
|Ncharged | 0 to 14| 2 to 14    |
